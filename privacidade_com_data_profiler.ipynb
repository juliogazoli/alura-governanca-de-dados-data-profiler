```python
# Importando bibliotecas
from dataprofiler import Data, Profiler
import pandas as pd
import json
```

```python
# Lendo os dados
dados = pd.read_csv('autoridades_navais.csv', encoding='windows-1252')
```

```python
dados.shape
```

```python
dados.head()
```

```python
# Gerando o profile
profile = Profiler(dados)
```

```python
# Gerando o Report
report = profile.report(
    report_options={'output_format': 'compact'}
)
```

```python
# Visualizando o report criado
print(json.dumps(report, indent=4))
```

```python
def verifica_dados_pessoais(results):
    colunas = []
    predicao = []
    exemplos = []

    for col in results['data_stats']:
        colunas.append(col['column_name'])
        predicao.append(col['data_label'])
        exemplos.append(col['samples'])

    df_results = pd.DataFrame({
        'Colunas': colunas,
        'Predicao': predicao,
        'Exemplos': exemplos,
    })

    return df_results
```

```python
results = verifica_dados_pessoais(report)
```
